```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('/content/heart.csv')
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['target'].value_counts()
```

```python
X = data.drop(columns='target',axis = 1)
Y = data['target']
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

**Model Selection**

#**Comparing the models with default hyperparameter values using Cross Validation**

```python
models =(LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier(random_state=0))
```

```python
def compare_models_cross_validation():
  for model in models:
    cv_score = cross_val_score(model, X, Y, cv = 5)
    mean_accuracy = sum(cv_score)/len(cv_score)
    mean_accuracy = round(mean_accuracy*100, 2)

    print('Cross Validation accuracies for the',model,'=', cv_score)
    print('Acccuracy score of the ',model,'=',mean_accuracy,'%')
    print('')
```

```python
compare_models_cross_validation()
```

**Random Forest Classifier** has the highest accuracy value with default hyperparameter values.

#**Comparing the models with different Hyperparameter values using GridSearchCV**

```python
models_list = [LogisticRegression(solver='saga',max_iter=200000), SVC(), KNeighborsClassifier(), RandomForestClassifier(random_state=0)]
```

```python
model_hyperparameters={
    'log_regression_hyperparameters':{
        'C':[1,5,10,20]
    },
    'svc_hyperparameters':{
        'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
        'C':[1,5,10,20]
    },
  'KNN_hyperparameters':{
      'n_neighbors':[3,5,10]
  },
  'random_forest_hyperparameters':{
      'n_estimators':[10,20,50,100]
  }
}
```

```python
model_keys = list(model_hyperparameters.keys())
```

```python
def ModelSelection(lists_of_models,hyperparameters_dictionary):
  result = []
  i = 0
  for model in lists_of_models:
    Key = model_keys[i]
    params = hyperparameters_dictionary[Key]
    i+=1
    print(model)
    print(params)
    print('')
    classifier = GridSearchCV(model, params, cv = 5)
    classifier.fit(X,Y)
    result.append({
        'model used' : model,
        'highest score' : classifier.best_score_,
        'best hyperparameters' : classifier.best_params_
    })
  result_dataframe = pd.DataFrame(result,columns=['model used','highest score','best hyperparameters'])
  return result_dataframe
```

```python
import warnings
warnings.filterwarnings("ignore",category=UserWarning)
```

```python
ModelSelection(models,model_hyperparameters)
```

**Random Forest Clssifier** with n_estimator = 100 has the highest accuracy.
