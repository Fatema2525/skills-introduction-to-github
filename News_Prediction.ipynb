```python
import re
import pandas as pd
import numpy as np
import nltk

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

from nltk.stem import PorterStemmer
```

```python
news=pd.read_csv('/content/NEWS.csv')
news.head()
```

```python
X=news['title'].astype(str).tolist()
Y=news['label'].values
```

```python
stemmer=PorterStemmer()
```

```python
def stem_text(text):
  text=text.lower()
  text=re.sub(r'[^a-z0-9\s]','',text)
  words=text.split()
  return [stemmer.stem(word) for word in words]
```

```python
X=[doc for doc in X if doc.strip()!=""]
```

```python
vectorizer=TfidfVectorizer(stop_words='english')
Z=vectorizer.fit_transform(X)
print(Z)
```

```python
Z_train,Z_test,Y_train,Y_test=train_test_split(Z,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model=LogisticRegression()
model.fit(Z_train,Y_train)
```

```python
Z_train_prediction=model.predict(Z_train)
training_data_acuuracy=accuracy_score(Z_train_prediction,Y_train)
```

```python
Z_test_prediction=model.predict(Z_test)
test_data_accuracy=accuracy_score(Z_test_prediction,Y_test)
```

```python
print("Training dat accuracy:",training_data_acuuracy)
print("Test data accuracy:",test_data_accuracy)
```

Prediction


```python
Z_new=Z_test[0]

prediction=model.predict(Z_new)

print(prediction)

if(prediction==0):
  print("The news is fake")
else:
  print("The news is real")
```
