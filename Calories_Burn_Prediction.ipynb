Importing The Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Data Collection and Processing


```python
data_1=pd.read_csv('/content/calories.csv')
```

```python
data_2=pd.read_csv('/content/exercise.csv')
```

```python
data=pd.concat([data_2,data_1],axis=1)
```

Converting Categorical Data

```python
data.replace({'Gender':{'male':0,'female':1}},inplace=True)
```

```python
data.head()
```

Separeting Features and Target data

```python
X=data.drop(columns=['User_ID','Calories'], axis=1)
Y=data['Calories']
```

```python
print(X)
```

```python
print(Y)
```

Splitting Training and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
model=XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

Prediction

```python
X_test_prediction=model.predict(X_test)
print(X_test_prediction)
```

Mean Absulate Error

```python
mae=metrics.mean_absolute_error(Y_test,X_test_prediction)
print('Mean Absulate Error:',mae)
```
